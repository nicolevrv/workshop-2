```python
#import libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pathlib import Path
```

```python

# Define the base path to the project directory with pathlib
base_path = Path.cwd().parent

spotify_path = base_path / "data" / "raw" / "spotify_dataset.csv"
grammy_path = base_path / "data" / "raw" / "the_grammy_awards.csv"

# Ensure the files exist before trying to read them
print(spotify_path.exists())
print(grammy_path.exists())
```

```python
#read spotify data
spotify_df = pd.read_csv(spotify_path)
spotify_df.head()
```

```python
#read grammy data
grammy_df = pd.read_csv(grammy_path)
grammy_df.head()
```

```python
#check for duplicates in spotify dataset
print("duplicates in spotify dataset:", spotify_df.duplicated().sum())
print("unique artists in spotify dataset:", spotify_df['artists'].nunique())
print("unique tracks in spotify dataset:", spotify_df['track_name'].nunique())
print("duplicates in spotify dataset (based on track ID):", spotify_df.duplicated(subset=['track_id']).sum())
print("duplicates in spotify dataset (based on track name and artist):", spotify_df.duplicated(subset=['track_name', 'artists']).sum())
print("duplicates in spotify dataset (based on track name, artist, and album name):", spotify_df.duplicated(subset=['track_name', 'artists', 'album_name']).sum())
print("duplicates in spotify dataset (based on track name, artist, album name, and track genre):", spotify_df.duplicated(subset=['track_name', 'artists', 'album_name', 'track_genre']).sum())
```

```python
#checking for non-logical duplicates in the spotify dataset based on track name, artist, album name, and track genre

duplicates_df = spotify_df[
    spotify_df.duplicated(
        subset=['track_name', 'artists', 'album_name', 'track_genre'],
        keep=False
    )
]


duplicates_df = duplicates_df.sort_values(
    by=['track_name', 'artists', 'album_name', 'track_genre']
)

duplicates_df.head(20)
```

```python
sns.histplot(spotify_df['popularity'], bins=30)
plt.title("Distribution of Track Popularity")
plt.xlabel("Popularity")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(spotify_df[spotify_df['popularity'] > 0]['popularity'], bins=30)
plt.title("Popularity Distribution (excluding zeros)")
plt.show()
```

```python
top_artists = (
    spotify_df['artists']
    .value_counts()
    .head(10)
)

sns.barplot(x=top_artists.values, y=top_artists.index)
plt.title("Top 10 Artists by Number of Tracks")
plt.xlabel("Number of Tracks")
plt.ylabel("Artist")
plt.show()
```

```python
top_genres = (
    spotify_df['track_genre']
    .value_counts()
    .head(10)
)

sns.barplot(x=top_genres.values, y=top_genres.index)
plt.title("Top 10 Genres")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.show()
```

```python
top_genres_list = spotify_df['track_genre'].value_counts().head(10).index

filtered_df = spotify_df[spotify_df['track_genre'].isin(top_genres_list)]

sns.boxplot(data=filtered_df, x='track_genre', y='popularity')
plt.xticks(rotation=45)
plt.title("Popularity by Genre")
plt.show()
```

```python
top_artist_popularity = (
    spotify_df.groupby('artists')['popularity']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

sns.barplot(x=top_artist_popularity.values, y=top_artist_popularity.index)
plt.title("Top Artists by Average Popularity")
plt.xlabel("Average Popularity")
plt.ylabel("Artist")
plt.show()
```

```python
grammy_df['year'].value_counts().sort_index().plot(kind='line')
plt.title("Number of Grammy Records per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```
